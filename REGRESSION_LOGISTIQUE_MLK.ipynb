```python
from scipy.stats import chi2_contingency
# SÃ©lectionner les variables catÃ©gorielles
cat_vars = ['Ct', 'Ft', 'Fm']

# Initialiser les listes pour stocker les rÃ©sultats
var_names = []
chi2_stats = []
p_values = []
cramer_vs = []

# Parcourir toutes les variables catÃ©gorielles
for var in cat_vars:
    # Calculer le tableau de contingence
    contingency_table = pd.crosstab(df_class['Pollution_labels'], df_class[var])
    # Calculer la statistique de test du Chi-deux et la p-valeur
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    # Calculer le coefficient V de Cramer
    n = contingency_table.sum().sum()
    phi2 = chi2/n
    r,k = contingency_table.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rc = r-((r-1)**2)/(n-1)
    kc = k-((k-1)**2)/(n-1)
    cramer_v = np.sqrt(phi2corr/min(rc-1,kc-1))
    # Ajouter les rÃ©sultats aux listes correspondantes
    var_names.append(var)
    chi2_stats.append(chi2)
    p_values.append(p)
    cramer_vs.append(cramer_v)

# CrÃ©er un DataFrame avec les rÃ©sultats
results_df = pd.DataFrame({
    'Variable': var_names,
    'Chi2': chi2_stats,
    'P-valeur': p_values,
    'V de Cramer': cramer_vs
})

# Trier le DataFrame par ordre croissant de V de Cramer
results_df.sort_values(by='V de Cramer', inplace=True,ascending=False)

# Afficher le tableau des rÃ©sultats
print(results_df)
```

```python
# GÃ©nÃ©rer un box plot pour chaque variable numÃ©rique
for var in num_vars:
    sns.boxplot(x='Pollution_labels', y=var, data=df_class)
    plt.title(var)
    plt.xlabel('Reponse binaire')
    plt.ylabel('Valeur')
    plt.show()
```

```python
import scipy.stats as stats
# Initialiser les listes pour stocker les rÃ©sultats
var_names = []
kw_stats = []
p_values = []

# Parcourir toutes les variables numÃ©riques
for var in num_vars:
    # Calculer les groupes de valeurs
    groups = [df_class[df_class['Pollution_labels'] == 0][var], df_class[df_class['Pollution_labels'] == 1][var]]
    # Appliquer le test de Kruskal-Wallis
    kw_stat, p = stats.kruskal(*groups)
    # Ajouter les rÃ©sultats aux listes correspondantes
    var_names.append(var)
    kw_stats.append(kw_stat)
    p_values.append(p)

# CrÃ©er un DataFrame avec les rÃ©sultats
results_df = pd.DataFrame({
    'Variable': var_names,
    'Kruskal-Wallis': kw_stats,
    'P-valeur': p_values
})

# Trier le DataFrame par ordre croissant de p-valeur
results_df.sort_values(by='P-valeur', inplace=True)

# Afficher le tableau des rÃ©sultats
print(results_df)
```

## II - InterpretabilitÃ© locale
L'interprÃ©tabilitÃ© locale des modÃ¨les consite en un ensemble de techniques destinÃ©es Ã  rÃ©pondre Ã  des questions telles ques :
- Pourquoi le modÃ¨le a -t- il fait cette prÃ©diction ?
- Quel a Ã©tÃ© l'impact de cette valeur de caracterisique spÃ©cifique sur la prÃ©diction?

## II - 1 LIME

LIME est une technique qui permet un modÃ¨le simple autour de la prÃ©diction que nous voulons expliquer et utilise ce modÃ¨le 
simple pour donner une explication . LIME peut Ãªtre utilisÃ© avec n'importe modÃ¨le de machine learning, pas seulement avec 
des arbres de dÃ©cision ou de forÃªts alÃ©atoires.

L'idÃ©e de base de LIME est assez simple. Pour une  prÃ©diction observÃ©e, LIME perturbe les entrÃ©es de la prÃ©diction et essaie 
de comprendre comment ces perturbations affectent la sortie du modÃ¨le. Les attributs qui affectent le plus la sortie lorqu'ils 
sont considÃ©rÃ©s comme ayant une importance Ã©lÃ©vÃ©e pour cette prÃ©diction spÃ©cifique.

## II - Shapley Value

La valeur du shapley , dans le contexte de l'interprÃ©tabilitÃ© de modÃ¨les d'apprentissage automatique, permet d'Ã©valuer 
l'impact de chaque caractÃ©ristique sur une prÃ©dition donnÃ©e
Elle fait cela en calculant la constribution moyenne de chaque caractÃ©ristique Ã  la prÃ©diction sur toutes les combinaisons 
possibles de caractÃ©ristiques

```python
# pip install shap
```

```python
import shap

# CrÃ©er un explainer SHAP pour le modÃ¨le de rÃ©gression logistique 
explainer = shap.LinearExplainer(clf, X_train)

# Calcul des valeurs SHAP pour les instances de test
shap_values = explainer.shap_values(X_test)

# Visualiser les valeurs de Shapley pour une prÃ©diction individuelle
shap.initjs() 
shap.force_plot(explainer.expected_value, shap_values[6], X_test.iloc[6])
```
